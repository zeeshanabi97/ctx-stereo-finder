```python
import geopandas as gp
from shapely.geometry import box, shape, Polygon
import ipyleaflet as ipyl
from ipyleaflet import Map, TileLayer, GeoData, DrawControl, WidgetControl, FullScreenControl, LayersControl
from ipyleaflet import projections
from ipywidgets import HTML, FloatText, Text, Button, VBox, HBox, Label, IntSlider, FloatSlider, Output, Dropdown, Checkbox
from IPython.display import display, clear_output
import numpy as np
import pandas as pd
import json
import time
import os

# Define the Mars CRS (Coordinate Reference Systems)
mars_eqc_crs = {'custom': True, 'name': 'Mars2000', 'proj4def': '+proj=eqc +lat_ts=0 +lat_0=0 +lon_0=0 +x_0=0 +y_0=0 +a=3396190 +b=3396190 +units=m +no_defs', }
mars_np_crs = {'custom': True, 'name': 'IAU2000:49919', 'proj4def': '+proj=stere +lat_0=90 +lon_0=0 +k=1 +x_0=0 +y_0=0 +a=3396190 +b=3376200 +units=m +no_defs', }
mars_sp_crs = {'custom': True, 'name': 'IAU2000:49920', 'proj4def': '+proj=stere +lat_0=-90 +lon_0=0 +k=1 +x_0=0 +y_0=0 +a=3396190 +b=3376200 +units=m +no_defs', }

# Basemap definitions
mola_color_shade_url = 'https://astro.arcgis.com/arcgis/rest/services/OnMars/MColorDEM/MapServer/tile/{z}/{y}/{x}'
mola_color_bm = {
    'url': mola_color_shade_url,
    'attribution': 'USGS/ESRI/NASA',
    'crs': projections.EPSG4326,
    'max_native_zoom': 5,
}
murray_ctx_global_mosaic = 'https://astro.arcgis.com/arcgis/rest/services/OnMars/CTX/MapServer/tile/{z}/{y}/{x}'
murray_ctx_global_mosaic_bm = {
    'url': murray_ctx_global_mosaic,
    'attribution': 'CalTech/USGS/MSSS/NASA',
    'crs': projections.EPSG4326,
    'max_native_zoom': 12,
}

# Define useful columns
good_columns = ['ProductId','Ext2URL', 'EmAngle', 'InAngle', 'PhAngle']
good_query_columns = ['ProductId_1', 'ProductId_2', 'EmAngle_1', 'EmAngle_2', 'diff_em', 'ovarea', 'Ext2URL_1', 'Ext2URL_2']

class CTXStereoFinder:
    def __init__(self):
        self.ctx_df = None
        self.stereo_results = None
        self.map_display = None
        self.output = Output()
        self.draw_control = None
        self.querys = None
        self.user_drawn_polygon = None
        self.create_ui()
        
    def create_ui(self):
        # Create UI elements
        self.center_lat = FloatText(value=0.0, description='Center Lat:', min=-90, max=90)
        self.center_lon = FloatText(value=0.0, description='Center Lon:', min=-180, max=360)
        self.box_size = FloatText(value=0.5, description='Box Size (°):', min=0.1, max=5.0)
        
        # Coordinate system options
        self.lon_format = Dropdown(
            options=[('0 to 360°', '0-360'), ('-180 to 180°', '-180-180')],
            value='0-360',
            description='Longitude Format:',
            style={'description_width': 'initial'}
        )
        
        self.min_emission_angle = FloatSlider(
            value=5.0, min=0.0, max=30.0, step=0.5,
            description='Min Emission Angle (°):',
            style={'description_width': 'initial'}
        )
        
        self.min_angle_diff = FloatSlider(
            value=10.0, min=1.0, max=30.0, step=0.5,
            description='Min Angle Diff (°):',
            style={'description_width': 'initial'}
        )
        
        self.max_angle_diff = FloatSlider(
            value=25.0, min=5.0, max=45.0, step=0.5,
            description='Max Angle Diff (°):',
            style={'description_width': 'initial'}
        )
        
        self.max_footprints = IntSlider(
            value=500, min=10, max=15000, step=10,
            description='Max Footprints:',
            style={'description_width': 'initial'}
        )
        
        self.use_cached_data = Checkbox(
            value=True,
            description='Use cached CTX data if available',
            style={'description_width': 'initial'}
        )
        
        # Load data button
        self.load_data_btn = Button(description='Load CTX Data')
        self.load_data_btn.on_click(self.on_load_data)
        
        # Search button
        self.search_btn = Button(description='Search Stereo Pairs', disabled=True)
        self.search_btn.on_click(self.on_search)
        
        # Export button
        self.export_btn = Button(description='Export Results', disabled=True)
        self.export_btn.on_click(self.on_export)
        
        # Layout
        coord_box = HBox([self.center_lat, self.center_lon, self.box_size])
        coord_options = HBox([self.lon_format, self.use_cached_data])
        param_box1 = HBox([self.min_emission_angle, self.max_footprints])
        param_box2 = HBox([self.min_angle_diff, self.max_angle_diff])
        buttons_box = HBox([self.load_data_btn, self.search_btn, self.export_btn])
        
        # Main UI
        main_ui = VBox([
            Label('Mars CTX Stereo Pair Finder'),
            Label('Step 1: Enter coordinates and region size'),
            coord_box,
            coord_options,
            Label('Step 2: Set stereo parameters'),
            param_box1, 
            param_box2,
            buttons_box,
            self.output
        ])
        
        display(main_ui)
        
        # Initialize the map for drawing
        self.initialize_map()
    
    def initialize_map(self):
        """Initialize the map with a draw control for selecting regions"""
        with self.output:
            # Create a map centered at the initial coordinates
            m = Map(
                center=(self.center_lat.value, self.center_lon.value),
                zoom=4,
                basemap=TileLayer(**mola_color_bm),
                crs=projections.EPSG4326,
            )
            m.add_layer(TileLayer(**murray_ctx_global_mosaic_bm, name='CTX Global Mosaic'))
            
            # Add a draw control for selecting regions
            draw_control = DrawControl()
            draw_control.rectangle = {
                "shapeOptions": {
                    "color": "#ff0000",
                    "fillOpacity": 0.2
                }
            }
            draw_control.polygon = {
                "shapeOptions": {
                    "color": "#ff0000",
                    "fillOpacity": 0.2
                }
            }
            
            # Handle draw events
            def handle_draw(self, action, geo_json):
                # Store the drawn polygon
                self.user_drawn_polygon = geo_json
                # Update the center coordinates based on the polygon
                if geo_json['geometry']['type'] == 'Polygon':
                    coords = geo_json['geometry']['coordinates'][0]
                    lons = [coord[0] for coord in coords]
                    lats = [coord[1] for coord in coords]
                    self.center_lat.value = np.mean(lats)
                    self.center_lon.value = np.mean(lons)
                    # Estimate box size
                    max_lat, min_lat = max(lats), min(lats)
                    max_lon, min_lon = max(lons), min(lons)
                    size = max(max_lat - min_lat, max_lon - min_lon)
                    self.box_size.value = size
            
            # Connect the handler
            draw_control.on_draw(handle_draw)
            m.add_control(draw_control)
            self.draw_control = draw_control
            
            # Add other controls
            m.add_control(FullScreenControl())
            m.add_control(LayersControl(position='bottomleft'))
            
            # Store and display the map
            self.map_display = m
            display(m)
            
            print("You can either enter coordinates manually or draw a region on the map.")
    
    def normalize_longitude(self, lon):
        """Normalize longitude to 0-360 range"""
        if self.lon_format.value == '0-360':
            # Ensure longitude is in 0-360 range
            while lon < 0:
                lon += 360
            while lon > 360:
                lon -= 360
        else:
            # Ensure longitude is in -180 to 180 range
            while lon < -180:
                lon += 360
            while lon > 180:
                lon -= 360
        return lon
    
    def on_load_data(self, b):
        """Handle loading CTX data"""
        with self.output:
            clear_output(wait=True)
            
            # Check if we have cached data
            cache_file = 'ctx_data_cache.geojson'
            if self.use_cached_data.value and os.path.exists(cache_file):
                print(f"Loading CTX footprints from cache...")
                try:
                    self.ctx_df = gp.read_file(cache_file)
                    print(f"Successfully loaded {len(self.ctx_df)} CTX footprints from cache")
                    self.search_btn.disabled = False
                    return
                except Exception as e:
                    print(f"Error loading cached data: {e}")
                    print("Will download data from source...")
            
            print("Loading CTX footprints data from USGS ODE (this may take a few minutes)...")
            try:
                ctx_fp_url = 'https://ode.rsl.wustl.edu/mars/datafile/derived_products/coverageshapefiles/mars/mro/ctx/edr/mars_mro_ctx_edr_c0a.zip'
                self.ctx_df = gp.read_file(ctx_fp_url)
                print(f"Successfully loaded {len(self.ctx_df)} CTX footprints")
                
                # Save to cache for future use
                try:
                    self.ctx_df.to_file(cache_file, driver='GeoJSON')
                    print(f"Saved data to cache for future use")
                except Exception as e:
                    print(f"Could not save cache: {e}")
                
                self.search_btn.disabled = False
            except Exception as e:
                print(f"Error loading CTX data: {e}")
                print("Detailed error information:")
                import traceback
                traceback.print_exc()
    
    def on_search(self, b):
        """Handle search button click"""
        with self.output:
            clear_output(wait=True)
            
            # Check if CTX data is loaded
            if self.ctx_df is None:
                print("Please load CTX data first")
                return
            
            # Use drawn polygon if available, otherwise use coordinates
            if self.user_drawn_polygon is not None:
                print("Using drawn polygon for search")
                self.querys = self.user_drawn_polygon
            else:
                # Create the search query for the specified region
                half_size = self.box_size.value / 2
                lat_min = self.center_lat.value - half_size
                lat_max = self.center_lat.value + half_size
                
                # Normalize longitudes based on selected format
                lon_center = self.normalize_longitude(self.center_lon.value)
                lon_min = self.normalize_longitude(lon_center - half_size)
                lon_max = self.normalize_longitude(lon_center + half_size)
                
                # Handle case where box crosses the prime meridian
                if lon_min > lon_max:
                    print("Warning: Search box crosses the prime meridian/date line")
                    print("Performing two separate searches and combining results")
                    
                    # First region: lon_min to 360 or 180 (depending on format)
                    lon_edge = 360 if self.lon_format.value == '0-360' else 180
                    query1 = self.build_query(lon_min, lon_edge, lat_min, lat_max)
                    
                    # Second region: 0 or -180 to lon_max
                    lon_start = 0 if self.lon_format.value == '0-360' else -180
                    query2 = self.build_query(lon_start, lon_max, lat_min, lat_max)
                    
                    # Combine queries
                    results1 = self.perform_stereo_query(query1)
                    results2 = self.perform_stereo_query(query2)
                    
                    if len(results1) > 0 and len(results2) > 0:
                        # Combine results
                        self.stereo_results = pd.concat([results1, results2])
                    elif len(results1) > 0:
                        self.stereo_results = results1
                    elif len(results2) > 0:
                        self.stereo_results = results2
                    else:
                        self.stereo_results = gp.GeoDataFrame(columns=[*good_query_columns, 'geometry'])
                    
                    # Display results
                    self.display_results()
                    return
                else:
                    # Standard case - box doesn't cross meridian
                    print(f"Searching for stereo pairs in region centered at {self.center_lat.value}°N, {lon_center}°E...")
                    self.querys = self.build_query(lon_min, lon_max, lat_min, lat_max)
            
            # Perform the stereo query
            self.stereo_results = self.perform_stereo_query(self.querys)
            
            # Display results
            self.display_results()
    
    def build_query(self, lon_min, lon_max, lat_min, lat_max):
        """Build a query for the given bounds"""
        return {
            'type': 'FeatureCollection',
            'features': [{
                'type': 'Feature',
                'geometry': {
                    'type': 'Polygon',
                    'coordinates': [[[lon_min, lat_min],
                                  [lon_min, lat_max],
                                  [lon_max, lat_max],
                                  [lon_max, lat_min],
                                  [lon_min, lat_min]]]
                },
                'properties': {}
            }],
        }
    
    def perform_stereo_query(self, querys):
        """Perform the stereo query with the given parameters"""
        try:
            # Convert query to GeoDataFrame
            try:
                query_gdf = gp.GeoDataFrame.from_features(querys['features'])
            except Exception as e:
                print(f"Error converting query to GeoDataFrame: {e}")
                return gp.GeoDataFrame(columns=[*good_query_columns, 'geometry'])
            
            # Get query bounds
            query_geom = query_gdf.geometry.iloc[0]
            minx, miny, maxx, maxy = query_geom.bounds
            
            print(f"Searching region with bounds: {minx:.2f}° to {maxx:.2f}° longitude, {miny:.2f}° to {maxy:.2f}° latitude")
            
            # Special handling for regions near the prime meridian/date line
            spans_meridian = False
            if abs(minx - maxx) > 180:
                spans_meridian = True
                print("Region spans the prime meridian/date line - performing special handling")
            
            # Filter by bounding box
            if spans_meridian:
                # For regions crossing the date line/prime meridian, we need special handling
                # Convert minx, maxx to 0-360 range
                minx_360 = minx % 360
                maxx_360 = maxx % 360
                
                # Get footprints on both sides
                df_left = self.ctx_df[(self.ctx_df.geometry.bounds.minx >= minx_360) | 
                                      (self.ctx_df.geometry.bounds.maxx <= maxx_360)].copy()
                
                # Also include footprints that cross the meridian
                df_cross = self.ctx_df[(self.ctx_df.geometry.bounds.minx <= minx_360) & 
                                       (self.ctx_df.geometry.bounds.maxx >= maxx_360)].copy()
                
                df_fp1 = pd.concat([df_left, df_cross])
            else:
                # Normal case - direct spatial filtering
                try:
                    df_fp1 = self.ctx_df.cx[minx:maxx, miny:maxy].copy()
                except Exception as e:
                    print(f"Error in spatial filtering: {e}")
                    df_fp1 = self.ctx_df[self.ctx_df.intersects(query_geom)].copy()
            
            # Debug output
            print(f"Initial filtering found {len(df_fp1)} CTX footprints in region")
            
            # If no footprints found, try alternative approach
            if len(df_fp1) == 0:
                print("No footprints found using spatial index. Trying direct geometric filtering...")
                df_fp1 = self.ctx_df[self.ctx_df.intersects(query_geom)].copy()
                print(f"Geometric filtering found {len(df_fp1)} CTX footprints")
            
            # Limit number of footprints
            max_footprints = self.max_footprints.value
            if len(df_fp1) > max_footprints:
                print(f"Found {len(df_fp1)} footprints, limiting to {max_footprints}")
                df_fp1 = df_fp1.iloc[0:max_footprints]
            
            # If still no footprints, return empty result
            if len(df_fp1) == 0:
                print("No CTX footprints found in region. Try a different area or larger box size.")
                return gp.GeoDataFrame(columns=[*good_query_columns, 'geometry'])
            
            # Use user-defined emission angle threshold
            min_em_angle = self.min_emission_angle.value
            
            # Perform query - divide into left and right images
            print(f"Finding stereo pairs using emission angle threshold of {min_em_angle}°")
            left = df_fp1[df_fp1['EmAngle'] < min_em_angle].copy()
            left.loc[:,'lg'] = left['geometry']
            right = df_fp1[df_fp1['EmAngle'] >= min_em_angle].copy()
            right.loc[:,'rg'] = right['geometry']
            
            # Skip if either left or right is empty
            if len(left) == 0 or len(right) == 0:
                print(f"Not enough images with emission angles below and above {min_em_angle}°")
                print(f"Found {len(left)} images with EmAngle < {min_em_angle}° and {len(right)} images with EmAngle >= {min_em_angle}°")
                print("Try adjusting the minimum emission angle threshold")
                return gp.GeoDataFrame(columns=[*good_query_columns, 'geometry'])
            
            print(f"Found {len(left)} potential left images and {len(right)} potential right images")
            
            # Perform overlay to find intersections
            try:
                print("Finding overlapping image pairs...")
                df_fp2 = gp.overlay(left, right, how='intersection')
                print(f"Found {len(df_fp2)} raw intersections")
                
                # Calculate emission angle difference and overlap area
                df_fp2['diff_em'] = (df_fp2['EmAngle_2'] - df_fp2['EmAngle_1']).abs() 
                df_fp2['ovarea'] = df_fp2['geometry'].to_crs(mars_eqc_crs['proj4def']).area
                
                # Clean up
                del df_fp2['lg']
                del df_fp2['rg']
                
                print(f"Found {len(df_fp2)} potential stereo pairs")
                return df_fp2
            except Exception as e:
                print(f"Error in overlay process: {e}")
                import traceback
                traceback.print_exc()
                return gp.GeoDataFrame(columns=[*good_query_columns, 'geometry'])
        except Exception as e:
            print(f"Unexpected error in stereo query: {e}")
            import traceback
            traceback.print_exc()
            return gp.GeoDataFrame(columns=[*good_query_columns, 'geometry'])
    
    def display_results(self):
        """Display the results of the search"""
        # Enable export button if we have results
        self.export_btn.disabled = len(self.stereo_results) == 0
        
        # Check if we found any results
        if len(self.stereo_results) > 0:
            # Sort by ideal stereo conditions
            sorted_results = self.stereo_results.sort_values(by=['ovarea', 'diff_em'], ascending=[False, False])
            
            # Get user-defined angle difference thresholds
            min_diff = self.min_angle_diff.value
            max_diff = self.max_angle_diff.value
            
            # Filter for ideal stereo angle differences
            ideal_stereo = sorted_results[(sorted_results['diff_em'] >= min_diff) & 
                                          (sorted_results['diff_em'] <= max_diff)]
            
            if len(ideal_stereo) > 0:
                print(f"Found {len(ideal_stereo)} ideal stereo pairs ({min_diff}-{max_diff}° difference)")
                display(ideal_stereo[good_query_columns].head(10))
            else:
                print(f"Found {len(sorted_results)} potential stereo pairs, but none in ideal range ({min_diff}-{max_diff}°)")
                print("Try adjusting the min/max angle difference parameters")
                display(sorted_results[good_query_columns].head(10))
            
            # Create and display map
            self.map_display = self.create_map()
            display(self.map_display)
            
            # Display text summary
            self.display_summary(ideal_stereo if len(ideal_stereo) > 0 else sorted_results)
        else:
            print("No stereo pairs found in this region")
            # Still show the map with the search region
            self.map_display = self.create_map()
            display(self.map_display)
    
    def create_map(self):
        """Create a map with the region of interest and stereo pair candidates"""
        # Get center coordinates
        center_lat = self.center_lat.value
        center_lon = self.normalize_longitude(self.center_lon.value)
        
        m = Map(
            center=(center_lat, center_lon),
            zoom=6, 
            basemap=TileLayer(**mola_color_bm),
            crs=projections.EPSG4326,
        )
        m.add_layer(TileLayer(**murray_ctx_global_mosaic_bm, name='CTX Global Mosaic'))
        
        # Add region of interest if we have one
        if self.querys is not None:
            roi = GeoData(
                geo_dataframe=gp.GeoDataFrame.from_features(self.querys['features']),
                name='Region of Interest',
                style={'color': 'red', 'fillOpacity': 0.1}
            )
            m.add_layer(roi)
        
        # Add stereo pairs if any found
        if self.stereo_results is not None and len(self.stereo_results) > 0:
            # Hover widget
            html = HTML('''Hover over a Stereo Candidate''')
            html.layout.margin = '0px 20px 20px 20px'
            hovercontrol = WidgetControl(widget=html, position='topright')
            m.add_control(hovercontrol)
            
            # Update HTML function
            def update_html(feature, **kwargs):
                html.value = '''
                    <h3>Emission Angle Diff: <b>{:.2f}°</b></h3>
                    <p>Overlap Area: <b>{:.2f} km²</b></p>
                    <a target="_blank" rel="noopener noreferrer" href="{}"><h4>Left Image: {}</h4></a>
                    <a target="_blank" rel="noopener noreferrer" href="{}"><h4>Right Image: {}</h4></a>
                '''.format(feature['properties']['diff_em'],
                       feature['properties']['ovarea']/1000000,
                       feature['properties']['Ext2URL_1'],
                       feature['properties']['ProductId_1'],
                       feature['properties']['Ext2URL_2'],
                       feature['properties']['ProductId_2'])
            
            # Add stereo results to map
            stereo_layer = GeoData(
                geo_dataframe=self.stereo_results[good_query_columns + ['geometry']],
                name='Stereo Candidates',
                hover_style={'fillColor': 'red', 'fillOpacity': 0.2},
            )
            stereo_layer.on_hover(update_html)
            m.add_layer(stereo_layer)
        
        # Add controls
        m.add_control(FullScreenControl())
        m.add_control(LayersControl(position='bottomleft'))
        
        return m
    
    def display_summary(self, best_pairs):
        """Display a text summary of the best stereo pairs"""
        min_diff = self.min_angle_diff.value
        max_diff = self.max_angle_diff.value
        
        # Filter again to make sure we have the right pairs
        filtered_pairs = best_pairs[(best_pairs['diff_em'] >= min_diff) & 
                                    (best_pairs['diff_em'] <= max_diff)]
        
        if len(filtered_pairs) > 0:
            print("\nTop 3 recommended stereo pairs:")
            for i, (idx, row) in enumerate(filtered_pairs.iloc[:3].iterrows()):
                print(f"\nPair {i+1}:")
                print(f"- Left image: {row['ProductId_1']} (Emission angle: {row['EmAngle_1']:.2f}°)")
                print(f"- Right image: {row['ProductId_2']} (Emission angle: {row['EmAngle_2']:.2f}°)")
                print(f"- Emission angle difference: {row['diff_em']:.2f}°")
                print(f"- Overlap area: {row['ovarea']/1000000:.2f} km²")
                print(f"- Left image URL: {row['Ext2URL_1']}")
                print(f"- Right image URL: {row['Ext2URL_2']}")
        else:
            print("\nNo stereo pairs match your criteria. Try adjusting the angle difference range.")
    
    def on_export(self, b):
        """Export the results to a file"""
        if self.stereo_results is None or len(self.stereo_results) == 0:
            with self.output:
                print("No results to export")
            return
        
        # Export to GeoJSON
        try:
            timestamp = time.strftime("%Y%m%d_%H%M%S")
            filename = f"ctx_stereo_results_{timestamp}.geojson"
            self.stereo_results.to_file(filename, driver='GeoJSON')
            
            # Also export as CSV for easier use
            csv_filename = f"ctx_stereo_results_{timestamp}.csv"
            self.stereo_results[good_query_columns].to_csv(csv_filename, index=False)
            
            with self.output:
                print(f"Exported results to {filename} and {csv_filename}")
        except Exception as e:
            with self.output:
                print(f"Error exporting results: {e}")

# Initialize the application
stereo_finder = CTXStereoFinder()
```
